```python
df_train = df_train.dropna(thresh=len(df_train) - int(len(df_train)/3), axis=1)
```

```python
df_train = df_train.dropna(axis=0)
```

```python
df_train_int = df_train.select_dtypes(exclude='object')
```

```python
def model_iter(X, y, model_list, test_frac):
    X_train, X_test, y_train, y_test = train_test_split(
     X, y, test_size=test_frac, random_state=42)
    scaler = StandardScaler()
    scaler.fit(X_train)
    X_train_s = scaler.transform(X_train)
    X_test_s = scaler.transform(X_test)
    for regr in model_list:
        regr_init = regr
        regr_init.fit(X_train_s, y_train)
        y_pred = regr_init.predict(X_test_s)
        print('Mean squared error for model {} is : {:.2f}\n'.format(regr,mean_squared_error(y_test, y_pred)))
```

```python
X = df_train_int.drop(columns='SalePrice')
y = df_train_int.SalePrice
model_list = [AdaBoostRegressor(), BaggingRegressor(), 
              RandomForestRegressor(n_estimators=100), GradientBoostingRegressor()]

model_iter(X,y,model_list,0.25)
```

```python
model_list = [AdaBoostRegressor(), BaggingRegressor(n_estimators=300), 
              RandomForestRegressor(n_estimators=300), GradientBoostingRegressor(n_estimators=300)]

model_iter(X,y,model_list,0.25)
```

```python
model_list = [GradientBoostingRegressor(n_estimators=10000)]

model_iter(X,y,model_list,0.15)
```
